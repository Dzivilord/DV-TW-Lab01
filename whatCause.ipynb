```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
from math import pi
```

* Vấn đề cần trực quan: Một số bệnh lý, thói quen xấu được cho nguyên nhân gây nên nguy cơ đau tim. Việc trực quan giúp kiểm tra xem đối với những người mắc nguy cơ đau tim có thói quen hay bệnh lý nào chiếm phổ biến nhất.
* Các biến sử dụng: Diabetes, Family History, Smoking, Obesity, Alcohol Consumption, Heart Attack Risk
* Loại biểu đồ: Biểu đồ radar
* Ý nghĩa của biểu đồ : Biểu đồ radar giúp đánh giá tần suất xuất hiện của các bệnh lý và thói quen xấu ở những người có nguy cơ đau tim. Với mỗi bệnh lý hoặc thói quen xấu, ta sẽ thấy mức độ phổ biến của nó (dựa trên tỷ lệ hoặc số lượng người mắc). Điều này giúp xác định yếu tố nào trong số các biến như tiểu đường, lịch sử gia đình, hút thuốc, béo phì, và tiêu thụ rượu có tác động mạnh nhất đến nguy cơ đau tim.
* Nguyên nhân sử dụng: việc biểu thị các bệnh lý, thói quen này trên radar chart sẽ dễ dáng cho thấy tác động của bệnh lý, thói quen này ảnh hưởng ít nhiều đến nguy cơ đau tim.

```python
df=pd.read_csv("dataset/data.csv")
```

```python


heart_attack_number = df[df['Heart Attack Risk'] == 1]

risk_factors = ['Diabetes', 'Family History', 'Smoking', 'Obesity', 'Alcohol Consumption']
factor_counts = heart_attack_number[risk_factors].sum()


total_people_at_risk = len(heart_attack_number)
factor_percentages = (factor_counts / total_people_at_risk) * 100


risks_list = list(risk_factors)  
values = factor_percentages.tolist()
values += values[:1]  


N = len(risks_list)

angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]


fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))


ax.fill(angles, values, color='yellow')
ax.plot(angles, values, color='black')


ax.set_xticks(angles[:-1])
ax.set_xticklabels(risks_list)


ax.set_yticklabels(['0%', '20%', '40%', '60%', '80%', '100%'])


plt.title('Risk Factors for Heart Attack', size=15, color='blue')
plt.show()
```

Kết luận: Nhìn chung hầu như tất cả bệnh lý, thói quen xấu đều xuất hiện ở những người mắc nguy cơ đau tim. Trong đó đặc biệt hút thuốc lá gây ảnh hưởng nặng nề nhất với gần 60% người mắc có thói quen này. May mắn thay nguy cơ mắc đau tim không di truyền hoàn toàn khi chỉ một phần nhỏ bệnh nhân có người thân bị tương tự.
