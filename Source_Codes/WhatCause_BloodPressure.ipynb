```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
from math import pi
```


# BIỂU ĐỒ RADAR: CÁC BỆNH LÝ, THÓI QUEN XẤU NÀO GÂY NÊN NGUY CƠ MẮC BỆNH TIM

1. **Các thuộc tính sử dụng**
- `Diabetes`: thuộc tính cho biết bệnh nhân có tiểu đường hay không (1: Có, 0: Không).
- `Family History`: thuộc tính cho biết người nhà bệnh nhân có tiền sử mắc bệnh tim hay không (1: Có, 0: Không).
- `Smoking`: thuộc tính cho biết bệnh nhân có hút thuốc hay không (1: Có, 0: Không).
- `Alcohol Consumption`: thuộc tính cho biết bệnh nhân có tiêu thụ cồn hay không (1: Có, 0: Không).
- `Heart Attack Risk`: bệnh nhân có nguy cơ mắc bệnh tim hay không (1: Có, 0: Không).
2. **Loại biểu đồ**

    **Biểu đồ Radar** : là một loại biểu đồ được sử dụng để trực quan hóa nhiều biến số (hoặc thuộc tính) của một đối tượng trong cùng một biểu đồ.
3. **Ý nghĩa của biểu đồ**

    Biểu đồ radar giúp đánh giá tần suất xuất hiện của các bệnh lý và thói quen xấu ở những người có nguy cơ đau tim. Với mỗi bệnh lý hoặc thói quen xấu, ta sẽ thấy mức độ phổ biến của nó (dựa trên tỷ lệ hoặc số lượng người mắc). Điều này giúp xác định yếu tố nào trong số các biến như tiểu đường, lịch sử gia đình, hút thuốc, béo phì, và tiêu thụ rượu có tác động mạnh nhất đến nguy cơ đau tim. 

4. **Nguyên nhân sử dụng**

    Việc biểu thị các bệnh lý, thói quen này trên radar chart sẽ dễ dáng cho thấy tác động của bệnh lý, thói quen này ảnh hưởng ít nhiều đến nguy cơ đau tim.

5. **Kết luận**
- Nhìn chung hầu như tất cả bệnh lý, thói quen xấu đều xuất hiện ở những người mắc nguy cơ đau tim. 
- Trong đó đặc biệt hút thuốc lá gây ảnh hưởng nặng nề nhất với gần 60% người mắc có thói quen này. May mắn thay nguy cơ mắc đau tim không di truyền hoàn toàn khi chỉ một phần nhỏ bệnh nhân có người thân bị tương tự.! 

```python

df=pd.read_csv("../Datasets/data.csv")
```

```python


heart_attack_number = df[df['Heart Attack Risk'] == 1]

risk_factors = ['Diabetes', 'Family History', 'Smoking', 'Obesity', 'Alcohol Consumption']
factor_counts = heart_attack_number[risk_factors].sum()


total_people_at_risk = len(heart_attack_number)
factor_percentages = (factor_counts / total_people_at_risk) * 100


risks_list = list(risk_factors)  
values = factor_percentages.tolist()
values += values[:1]  


N = len(risks_list)

angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]


fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))


ax.fill(angles, values, color='yellow')
ax.plot(angles, values, color='black')


ax.set_xticks(angles[:-1])
ax.set_xticklabels(risks_list)


ax.set_yticklabels(['0%', '20%', '40%', '60%', '80%', '100%'])


plt.title('Risk Factors for Heart Attack', size=15, color='blue')
plt.show()
```


# BIỂU ĐỒ ĐƯỜNG: HUYẾT ÁP CỦA MỘT NGƯỜI BỊ ẢNH HƯỞNG NHƯ THẾ NÀO BỞI THỜI GIAN NGỦ CỦA HỌ

1. **Các thuộc tính sử dụng**
- `Systolic blood pressure`: huyết áp tâm thu
- `Diastolic blood pressure`: huyết áp tâm trương
- `Sleep Hours Per Day`: thời gian ngủ của đối tượng trong một ngày.
2. **Loại biểu đồ**

    **Biểu đồ đường** : Biểu đồ đường (line chart) là một công cụ trực quan hóa dữ liệu được sử dụng để thể hiện sự thay đổi của một hoặc nhiều biến số theo thời gian hoặc theo một trục đo lường khác.
3. **Ý nghĩa của biểu đồ**

    Biểu đồ này hỗ trợ trực quan xem huyết áp trung bình của những người có cùng thời gian ngủ, với cột dọc là chỉ số huyết áp của tâm thu và huyết áp tâm trương, cột ngang thể hiện số giờ ngủ của họ. Huyết áp là một yếu tố quan trọng giúp xác định nguy cơ mắc bệnh tim của một người. Nếu thực sự thời gian ngủ có thể ảnh hưởng đến huyết áp thì con người có thể điều chỉnh thời gian ngủ hợp lý để tránh nguy cơ mắc bệnh tim. 

4. **Nguyên nhân sử dụng**

    Biểu đồ đường dễ dàng  giúp chúng ta xác định xu hướng của huyết áp khi thời gian ngủ thay đổi.Nếu thời gian ngủ càng ít hoặc càng nhiều dẫn đến thay đổi huyết áp (tăng hoặc giảm), biểu đồ đường sẽ làm nổi bật sự thay đổi này một cách rõ ràng.

5. **Kết luận**

    Nhìn chung thời gian ngủ có ảnh hưởng đến huyết áp nhưng không quá đáng kể. Đối với những người có thời gian ngủ ít thì huyết áp có sự biến đổi mạnh. Với những người có thời gian ngủ phù hợp, từ 6 tiếng trở lên thì huyết áp hầu như không thay đổi đáng kể.

```python
average_bp = df.groupby('Sleep Hours Per Day')[['Systolic blood pressure', 'Diastolic blood pressure']].mean()

average_bp.reset_index(inplace=True)

plt.figure(figsize=(8, 6))

plt.plot(average_bp['Sleep Hours Per Day'], average_bp['Systolic blood pressure'], label='Average Systolic BP', color='blue', marker='o')


plt.plot(average_bp['Sleep Hours Per Day'], average_bp['Diastolic blood pressure'], label='Average Diastolic BP', color='red', marker='o')
plt.ylim(80, 140)

plt.title('Effect of Sleep Hours on Average Blood Pressure')
plt.xlabel('Sleep Hours Per Day')
plt.ylabel('Blood Pressure (mm Hg)')

plt.legend()

plt.grid(True)

plt.show()
```
